## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import numpy as np


# Handling Missing Values
df.fillna({'children': 0, 'country': 'Unknown', 'agent': 0, 'company': 0}, inplace=True)
# Drop rows with missing values in important columns
df.dropna(subset=['hotel', 'arrival_date_year', 'arrival_date_month'], inplace=True)

# Data Cleaning
# Remove rows with incorrect values
df = df[df['adults'] + df['children'] + df['babies'] > 0]
df = df[df['is_canceled'].isin([0, 1])]

# Feature Engineering
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Convert date columns to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Remove unnecessary columns
columns_to_drop = ['arrival_date_year', 'agent', 'company']
df.drop(columns=columns_to_drop, inplace=True)

# Convert categorical variables to dummy/indicator variables
categorical_columns = ['hotel', 'arrival_date_month', 'meal', 'country', 'market_segment', 'distribution_channel', 'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type', 'reservation_status']
df = pd.get_dummies(df, columns=categorical_columns)

# Save the cleaned dataset
df.to_csv('hotel_booking_cleaned.csv', index=False)

print("Data wrangling completed and cleaned dataset saved.")
````

````python
# Chart - 1 visualization code

# Load the cleaned dataset
df = pd.read_csv('hotel_booking_cleaned.csv')

# Set the style for visualization
sns.set(style="whitegrid")

# Create a box plot to show the relationship between lead time and cancellations
plt.figure(figsize=(10, 6))
sns.boxplot(x='is_canceled', y='lead_time', data=df)
plt.title('Lead Time vs. Cancellations')
plt.xlabel('Booking Cancellation')
plt.ylabel('Lead Time (days)')
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose the **box plot** specifically for its ability to effectively display the relationship between a **categorical variable (booking cancellation)** and a **continuous variable (lead time)**.
Here's why the box plot is a suitable choice for this scenario:

- The box plot allows us to compare the distributions of lead time for both canceled and not-canceled bookings.
- This is useful for understanding whether there are differences in lead times between the two groups.

**2. Summary Statistics:**
- The box plot displays important summary statistics such as the median, quartiles, and potential outliers for each group.
- This provides a quick overview of the central tendency and spread of the data.

**3. Visualizing Spread:**
- The width of the boxes and whiskers visually represents the spread of the data.
- A wider spread for canceled bookings might indicate higher variability in lead times for those cases.

**4. Outliers Detection:**
- Outliers, if present, are clearly displayed beyond the whiskers of the box plot.
- This helps identify extreme values that could potentially impact the analysis.

**5. Categorical Comparison:**
- By using distinct boxes for each cancellation status (0 or 1), the chart provides a clear visual comparison of lead times between the two groups.

In this context, we understand how lead time varies based on booking cancellations. The box plot efficiently communicates this relationship, making it a suitable choice for exploring and explaining the relationship between these variables.
